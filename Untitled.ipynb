## steps1 : import all required models at first 

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## steps2: load the weights and config files to net object

```python
net = cv2.dnn.readNet("yolov3 (1).weights","yolov3.cfg")
```

```python
classes = []          # extract the objects names from coco files in classes list
with open('coco.names.txt','r') as f:
    classes = f.read().splitlines()
```

```python
classes
```

## step 3: load the image on which objects get detect and take the dimensions of the imgs 

```python
img = cv2.imread('01-test-image.jpg')
cv2.imshow('Image',img)
height,width,_ = img.shape
print(height,width)
```

```python
plt.imshow(img) 
```

## step 3: Blob from image or crop the objects from image

```python
blob = cv2.dnn.blobFromImage(img,1/255,(416,416),(0,0,0),swapRB = True,crop = False)  
# convert pgr to rgb and scale the model by dividing to 1/255 and size to 416 ,416
```

```python
net.setInput(blob)
```

```python
out_ly_na = net.getUnconnectedOutLayersNames()
layerout = net.forward(out_ly_na)
```

## step4: extrat the images with higher confidence 

```python
boxes = []
confidences = []
class_ids = []
for output in layerout:
    for detection in output:
        scores = detection[5:]
        class_id = np.argmax(scores)
        confidence = scores[class_id]  # confidence gives the probeblity of how the objects is predicted correctly
        if confidence >0.5:
            center_x,center_y,w,h = int(detection[0]*width),int(detection[1]*height),int(detection[2]*width),int(detection[3]*height)
            x = int(center_x -w/2)
            y = int(center_y - h/2)
            boxes.append([x,y,w,h])
            confidences.append((float(confidence)))
            class_ids.append(class_id)
```

```python
print(len(boxes))
```

```python
indexes = cv2.dnn.NMSBoxes(boxes,confidences,0.5,0.4)
print(sorted(indexes.flatten()))
```

## step5: set box the images with higher confidence 

```python
font = cv2.FONT_HERSHEY_PLAIN
colors = np.random.uniform(0,255,size = (len(boxes),3))
obj = {}
for i in indexes.flatten():
    x,y,w,h = boxes[i]
    label = str(classes[class_ids[i]])
    confidence = str(round(confidences[i],2))
    color = colors[i]
    cv2.rectangle(img,(x,y),(x+w,y+h),color,2)
    if label in obj:
        obj[label] += 1
    else:
        obj[label] = 1
    cv2.putText(img,label+" "+confidence,(x,y+20),font,2,(255,255,255),2)
```
